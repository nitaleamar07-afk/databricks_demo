```sql
use catalog join_practise
```

```sql
create schema emp_db
```

```sql
CREATE TABLE emp_db.employees (
  emp_id INT,
  emp_name STRING,
  dept_id INT
);
```

```sql
select * from emp_db.employees
```

```sql
create table emp_db.departments (
  dept_id INT,
  dept_name STRING
)
```

```sql
create table emp_db.salaries(
  salary INT,
  emp_id INT)
```

```sql
select emp_name,s.emp_id from emp_db.employees e left join emp_db.salaries s on e.emp_id = s.salary 
```

```sql
drop table emp_db.salaries
```

```sql
CREATE TABLE emp_db.salaries (
  emp_id INT,
  salary INT
);
```

```sql
INSERT INTO emp_db.salaries VALUES
(1, 60000),
(2, 50000),
(3, 70000);
```

```sql
select emp_name,s.salary from emp_db.employees e left join emp_db.salaries s on e.emp_id = s.emp_id 
```

```sql
select d.dept_name,e.emp_name, s.salary from emp_db.employees e left join emp_db.salaries s on e.emp_id = s.emp_id left join emp_db.departments d on e.dept_id = d.dept_id
```

```sql
select * from emp_db.employees as e inner join emp_db.departments as d on e.dept_id = d.dept_id inner join emp_db.salaries as s on e.emp_id = s.emp_id
```
